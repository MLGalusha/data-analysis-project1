```python
import pandas as pd
from dotenv import load_dotenv
import os
import time
import json
import requests
import matplotlib.pyplot as plt
```

```python
# Read the CSV files into DataFrames.
UnemploymentRate_df = pd.read_csv('unemployment_rate.csv')
```

```python
# Display the Unemployment Rate DataFrame
UnemploymentRate_df.head(15)
```

```python
# Check the Unemployment Rate data types.
UnemploymentRate_df.info()
```

```python
# Check if any values are null.
UnemploymentRate_df.isnull().sum()
```

```python
# Show the unemployment rate for every 20 years using groupby
UnemploymentRate_df['20YearPeriod'] = (UnemploymentRate_df['Date'] // 20) * 20
unemployment_by_period = UnemploymentRate_df.groupby('20YearPeriod')['Unemployment Rate(%)'].mean().reset_index()

# Rename columns
unemployment_by_period.columns = ['Start Year', 'Average Unemployment Rate']

print("\nAverage unemployment rate for each 20-year period:")
print(unemployment_by_period)
```

```python
# Bar chart that shows unemployment rate by year
plt.bar(UnemploymentRate_df['Date'], UnemploymentRate_df['Unemployment Rate(%)'], color='blue')


plt.title('Unemployment Rate by Year')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.xticks(rotation=0)


plt.tight_layout()
plt.show()
```

```python
# Bar graph showing employment to populatoin rate by year
plt.bar(UnemploymentRate_df['Date'], UnemploymentRate_df['Employment to Population Rate(%)'], color='green')


plt.title('Employment to Population Rate(%)')
plt.xlabel('Date')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Scatter plot showing comparison between unemployment rate and employment to populatoin rate by year
plt.scatter(UnemploymentRate_df['Date'], UnemploymentRate_df['Unemployment Rate(%)'], color='blue', alpha=0.7, label='Unemployment Rate')
plt.scatter(UnemploymentRate_df['Date'], UnemploymentRate_df['Employment to Population Rate(%)'], color='green', alpha=0.7, label='Employment to Population Rate')

plt.title('Unemployment Rate vs Employment to Population Rate')
plt.xlabel('Date')
plt.ylabel('Rate (%)')
plt.xticks(rotation=0)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
